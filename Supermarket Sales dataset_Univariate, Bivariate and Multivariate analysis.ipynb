- Display Daily Sales by Day of the week

```python
dw_mapping = {
    0:'Mon',
    1:'Tue',
    2:'Wed',
    3:'Thur',
    4:'Fri',
    5:'Sat',
    6:'Sun'
}
```

```python
data['day_of_week'] = data['Date'].dt.dayofweek.map(dw_mapping)
```

```python
data.head(1)
```

```python
data['day_of_week'].value_counts().plot(kind='bar')
```

- What will be the highest month for the sale

```python
# Create a python dictionary for month

month_mapping = {
    1:'Jan',
    2:'Feb',
    3:'March',
    4:'April',
    5:'May',
    6:'Jun',
    7:'Jul',
    8:'Aug',
    9:'Sep',
    10:'Oct',
    11:'Nov',
    12:'Dec'
}
```

```python
data['month'] = data['Date'].dt.month.map(month_mapping)
```

```python
data['month'].value_counts().plot(kind='bar')
```
